# Overview

https://leetcode.com/problems/combination-sum-ii/

Given a collection of candidate numbers (candidates) and a target number (target), find all unique combinations in candidates where the candidate numbers sum to target.

Each number in candidates may only be used once in the combination.

Note: The solution set must not contain duplicate combinations.

 

## Example 1:
```
Input: candidates = [10,1,2,7,6,1,5], target = 8
Output: 
[
[1,1,6],
[1,2,5],
[1,7],
[2,6]
]
```


## Example 2:

```
Input: candidates = [2,5,2,1,2], target = 5
Output: 
[
[1,2,2],
[5]
]
```

# Solution

````python
from typing import List
from collections import Counter
````

````python
class CombinationSum:
    '''
    
    `construct_candidates` should store indicices, not values.

    '''

    def __init__(self, data, target):
        self.data = sorted(data)

        self.target = target
        self.n = len(self.data)
        
#         print(f"self.n: {self.n}")
        self.res = None
    
    def is_a_solution(self, a, k, rem) -> bool:
        res = rem == 0

#         print(f"[{k}] is_a_solution rem: {rem} a:{a[:k+1]} -> {res}")
        
        # if True and res:
        #     a_sum = sum(map(lambda i: self.data[i], a[:k+1]))
        #     assert a_sum == self.target

        return res

    
    def process_solution(self, a, k):
        self.res.append(list(map(lambda i: self.data[i],a[:k+1])))

        
    def construct_candidates(self, a, k, rem):
        res = []
        last = a[k-1]

#         print(f"[{k}] 111 construct_candidates a:{a[:k]} last: {last}")
        if last is None:
            j = 0
        else:
            j = last + 1

        if j >= self.n:
            return res
        if (self.n-j == 1):
            res = [j]
        else:
            res.append(j)
            # print(f"[{k}] 2222 construct_candidates a:{a[:k]} last: {last} res: {res}")
            prev_v = self.data[j]
            for i in range(j+1,self.n):

                # early termonation as all variants are already sorted out
                if self.data[i] > rem:
                    break

                if self.data[i] == prev_v:
                    continue
                res.append(i)
                prev_v = self.data[i]
                
        # print(f"[{k}] construct_candidates a:{a[:k]} last: {last} -> {res}")

        return res
        
    def backtrack(self, a, k, rem: int):
        '''
        a - array of partial solutions. Contains indices only.
        '''
#         print(f"[{k}] backtrack rem:{rem} -> {a[:k+1]}")
        if self.is_a_solution(a, k, rem):
            self.process_solution(a, k)
        else:
            k += 1
            c = self.construct_candidates(a, k, rem)
            for cv in c:
                a[k] = cv
                self.backtrack(a, k, rem - self.data[cv])
    
    def get_result(self) -> List[List[int]]:
        self.res = []
        a = [None] * self.n
        self.backtrack(a, -1, self.target)
        return self.res

# CombinationSum([1], 1).get_result()    
# CombinationSum([1,2], 3).get_result()    
# CombinationSum([1], 1).get_result()    
# CombinationSum([1,1,2,3], 4).get_result()
# CombinationSum([10,1,2,7,6,1,5], 8).get_result()
# CombinationSum([1,2,1], 3).get_result()
# CombinationSum([1,1,2], 3).get_result()


CombinationSum([14,6,25,9,30,20,33,34,28,30,16,12,31,9,9,12,34,16,25,32,8,7,30,12,33,20,21,29,24,17,27,34,11,17,30,6,32,21,27,17,16,8,24,12,12,28,11,33,10,32,22,13,34,18,12],27).get_result()
````
