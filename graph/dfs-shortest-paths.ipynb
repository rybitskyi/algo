```python
gv('''
1->2->7->1
3->6->7->3
''')


# source = 1, target = 6
```

```python
g = defaultdict(set)
g

# (1, 0) (2, 1) (7, 2) (3, 3) (6, 4) found dist: 4
```

```python
# Route #1
g[1].add(2)
g[2].add(7)
g[7].add(1)
g
```

```python
# Route #2
g[3].add(6)
g[6].add(7)
g[7].add(3)

for r in g.items():
    print(r)
```

```python
def dfs():
    source = 1
    target = 6
    
    visited = set([source])
    q = deque([(source, 0)])
    while q:
        u, dist = q.pop()
        print((u, dist), end = ' ')
        if u == target:
            print(f'found dist: {dist}')
            return dist
        for v in g[u]:
            if v in visited:
                continue
            
            q.appendleft((v, dist + 1))

dfs()
```
